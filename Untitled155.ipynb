```python
import os
import pandas as pd
import numpy as np
import joblib
import json
import yaml
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score

# ------------------------------------------------------------
# 🧩 Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Irrigation & Water Optimization"
MODEL_PATH = os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_model.pkl")
SCALER_PATH = os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_scalers.pkl")
CONFIG_PATH = os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_config.yaml")
NEW_DATA_PATH = os.path.join(BASE_DIR, "archive", "plant_vase1.CSV")

# ------------------------------------------------------------
# 🧠 Load model, scalers, and config
# ------------------------------------------------------------
model = joblib.load(MODEL_PATH)
scalers = joblib.load(SCALER_PATH)
scaler_x, scaler_y = scalers["scaler_x"], scalers["scaler_y"]

with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)

selected_features = config.get("selected_features", [])
print("[INFO] Model, scalers, and config loaded.")
print(f"[INFO] Model trained with {len(selected_features)} features: {selected_features}")

# ------------------------------------------------------------
# 📂 Load new data
# ------------------------------------------------------------
df = pd.read_csv(NEW_DATA_PATH)
print("[INFO] Loaded new dataset:", df.shape)

df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df = df.dropna().reset_index(drop=True)

# ------------------------------------------------------------
# 🧩 Filter only AIS-selected features
# ------------------------------------------------------------
missing = [f for f in selected_features if f not in df.columns]
if missing:
    raise ValueError(f"❌ Missing expected features in dataset: {missing}")

X = df[selected_features].values

# Optional target column
target_col = "water_required" if "water_required" in df.columns else None
y = df[target_col].values if target_col else None

# ------------------------------------------------------------
# ⚙️ Handle scaler dimension mismatch
# ------------------------------------------------------------
if X.shape[1] != scaler_x.n_features_in_:
    print(f"[WARN] Scaler expects {scaler_x.n_features_in_} features, "
          f"but current input has {X.shape[1]}. Re-fitting scaler safely...")
    new_scaler_x = MinMaxScaler()
    X_scaled = new_scaler_x.fit_transform(X)
    scaler_x = new_scaler_x  # override safely for this session
else:
    X_scaled = scaler_x.transform(X)

# ------------------------------------------------------------
# 🔮 Predict
# ------------------------------------------------------------
y_pred_scaled = model.predict(X_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).flatten()
df["Predicted_Water_Requirement"] = y_pred

# ------------------------------------------------------------
# 📊 Evaluate (if actual available)
# ------------------------------------------------------------
if y is not None:
    y_actual = scaler_y.inverse_transform(y.reshape(-1, 1)).flatten()
    rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
    r2 = r2_score(y_actual, y_pred)
    print(f"[RESULT] RMSE={rmse:.4f} | R²={r2:.4f}")
else:
    rmse, r2 = None, None
    print("[INFO] No 'water_required' column found; skipping evaluation.")

# ------------------------------------------------------------
# 💾 Save outputs
# ------------------------------------------------------------
pred_csv = os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_prediction_results.csv")
summary_json = os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_prediction_summary.json")

df.to_csv(pred_csv, index=False)
summary = {
    "dataset_used": NEW_DATA_PATH,
    "num_records": len(df),
    "rmse": float(rmse) if rmse else None,
    "r2_score": float(r2) if r2 else None,
    "used_features": selected_features,
}
with open(summary_json, "w") as f:
    json.dump(summary, f, indent=4)

# ------------------------------------------------------------
# 📈 Visualization
# ------------------------------------------------------------
plt.figure(figsize=(9,5))
if y is not None:
    plt.plot(y_actual[:200], label="Actual", color="green", linewidth=2)
plt.plot(y_pred[:200], label="Predicted", color="blue", linestyle="--", linewidth=2)
plt.title("🔮 AIS+PSO Hybrid AquaHarvest — Prediction on New Data")
plt.xlabel("Sample Index")
plt.ylabel("Water Requirement")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_new_prediction_graph.png"))
plt.show()

if y is not None:
    errors = y_actual - y_pred
    plt.figure(figsize=(7,5))
    sns.histplot(errors, bins=30, kde=True, color="tomato")
    plt.title("📊 Error Distribution (New Dataset)")
    plt.xlabel("Prediction Error")
    plt.tight_layout()
    plt.savefig(os.path.join(BASE_DIR, "Ais_Pso_hybrid_AquaHarvest_new_error_graph.png"))
    plt.show()

print("\n✅ [INFO] Prediction and graphs saved successfully in:")
print(BASE_DIR)
print("---------------------------------------------------")
print(" - Ais_Pso_hybrid_AquaHarvest_prediction_results.csv")
print(" - Ais_Pso_hybrid_AquaHarvest_prediction_summary.json")
print(" - Ais_Pso_hybrid_AquaHarvest_new_prediction_graph.png")
if y is not None:
    print(" - Ais_Pso_hybrid_AquaHarvest_new_error_graph.png")
print("---------------------------------------------------")
```
