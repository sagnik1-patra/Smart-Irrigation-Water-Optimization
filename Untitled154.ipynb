```python
# ============================================================
# 🌿 AquaHarvest — Hybrid AIS + PSO Model for Smart Irrigation
# ============================================================

import os
import numpy as np
import pandas as pd
import json
import yaml
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# ------------------------------------------------------------
# 📁 Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Irrigation & Water Optimization"
FILES = [
    os.path.join(BASE_DIR, "archive", "plant_vase1(2).CSV"),
    os.path.join(BASE_DIR, "archive", "plant_vase1.CSV"),
    os.path.join(BASE_DIR, "archive", "plant_vase2.CSV")
]
os.makedirs(BASE_DIR, exist_ok=True)

# ------------------------------------------------------------
# 📊 Load & Merge
# ------------------------------------------------------------
dfs = [pd.read_csv(f) for f in FILES]
data = pd.concat(dfs, ignore_index=True)
print("[INFO] Loaded data:", data.shape)

data.columns = [c.strip().lower().replace(" ", "_") for c in data.columns]
data = data.dropna().reset_index(drop=True)

# ------------------------------------------------------------
# 🧩 Feature/Target setup
# ------------------------------------------------------------
if 'timestamp' in data.columns:
    data['timestamp'] = pd.to_datetime(data['timestamp'], errors='coerce')
    data = data.sort_values('timestamp')

feature_cols = [c for c in data.columns if c not in ['timestamp', 'target', 'water_required']]
target_col = 'water_required' if 'water_required' in data.columns else feature_cols[-1]

X = data[feature_cols].values
y = data[target_col].values

# Normalize
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# ============================================================
# 🧠 Step 1: Artificial Immune System (AIS) Feature Selection
# ============================================================
np.random.seed(42)
num_features = X_train.shape[1]
population_size = 10
num_generations = 8

def fitness(selected):
    selected_idx = np.where(selected == 1)[0]
    if len(selected_idx) == 0:
        return 1e9
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train[:, selected_idx], y_train)
    pred = model.predict(X_test[:, selected_idx])
    return mean_squared_error(y_test, pred)

population = np.random.randint(0, 2, (population_size, num_features))

for gen in range(num_generations):
    fitness_scores = np.array([fitness(ind) for ind in population])
    best_idx = np.argmin(fitness_scores)
    best_solution = population[best_idx].copy()
    for i in range(population_size):
        mutation_mask = np.random.rand(num_features) < 0.2
        clone = best_solution.copy()
        clone[mutation_mask] = 1 - clone[mutation_mask]
        population[i] = clone
    print(f"[AIS] Generation {gen+1}/{num_generations} | Best MSE={fitness_scores[best_idx]:.6f}")

best_features_idx = np.where(best_solution == 1)[0]
selected_features = [feature_cols[i] for i in best_features_idx]
print("[AIS] Selected Features:", selected_features)

X_train_sel = X_train[:, best_features_idx]
X_test_sel = X_test[:, best_features_idx]

# ============================================================
# 🐦 Step 2: Particle Swarm Optimization (PSO) for RF tuning
# ============================================================
num_particles = 8
max_iter = 10
w, c1, c2 = 0.7, 1.5, 1.5

param_min = np.array([50, 2])
param_max = np.array([300, 20])

pos = np.random.uniform(param_min, param_max, (num_particles, 2))
vel = np.zeros_like(pos)

pbest = pos.copy()
pbest_scores = np.array([1e9]*num_particles)
gbest = pos[0].copy()
gbest_score = 1e9

def evaluate(params):
    n_est = int(max(10, min(params[0], 500)))   # clamp 10–500
    depth = int(max(2, min(params[1], 50)))     # clamp 2–50
    model = RandomForestRegressor(n_estimators=n_est, max_depth=depth, random_state=0)
    model.fit(X_train_sel, y_train)
    pred = model.predict(X_test_sel)
    return mean_squared_error(y_test, pred)

for it in range(max_iter):
    for i in range(num_particles):
        mse = evaluate(pos[i])
        if mse < pbest_scores[i]:
            pbest_scores[i], pbest[i] = mse, pos[i].copy()
        if mse < gbest_score:
            gbest_score, gbest = mse, pos[i].copy()
    r1, r2 = np.random.rand(), np.random.rand()
    vel = w*vel + c1*r1*(pbest - pos) + c2*r2*(gbest - pos)
    pos += vel
    pos = np.clip(pos, param_min, param_max)
    print(f"[PSO] Iter {it+1}/{max_iter} | Best MSE={gbest_score:.6f}")

best_n_est = int(max(10, min(gbest[0], 500)))
best_depth = int(max(2, min(gbest[1], 50)))
print(f"[PSO] Best Params → n_estimators={best_n_est}, max_depth={best_depth}")

# ============================================================
# 🚀 Step 3: Train Final Hybrid Model
# ============================================================
model = RandomForestRegressor(n_estimators=best_n_est, max_depth=best_depth, random_state=0)
model.fit(X_train_sel, y_train)
pred_scaled = model.predict(X_test_sel)
pred = scaler_y.inverse_transform(pred_scaled.reshape(-1, 1)).flatten()
actual = scaler_y.inverse_transform(y_test.reshape(-1, 1)).flatten()

rmse = np.sqrt(mean_squared_error(actual, pred))
r2 = r2_score(actual, pred)
print(f"[RESULT] RMSE={rmse:.4f} | R²={r2:.4f}")

# ============================================================
# 📊 Step 4: Visualization
# ============================================================
OUTPUT_DIR = BASE_DIR

# 1️⃣ Heatmap
plt.figure(figsize=(8,6))
corr = data[selected_features + [target_col]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("AIS-Selected Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_heatmap.png"))
plt.show()

# 2️⃣ Prediction Graph
plt.figure(figsize=(9,5))
plt.plot(actual[:200], label="Actual", color='green', linewidth=2)
plt.plot(pred[:200], label="Predicted (Hybrid)", color='blue', linestyle='--', linewidth=2)
plt.title("🔮 Actual vs Predicted (AIS+PSO Hybrid)")
plt.xlabel("Sample Index")
plt.ylabel("Water Requirement")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_prediction_graph.png"))
plt.show()

# 3️⃣ Result Graph
errors = actual - pred
plt.figure(figsize=(7,5))
sns.histplot(errors, bins=30, kde=True, color="orange")
plt.title("📊 Error Distribution (AIS+PSO Hybrid)")
plt.xlabel("Prediction Error")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_result_graph.png"))
plt.show()

# 4️⃣ Comparison Graph
plt.figure(figsize=(9,5))
plt.plot(actual[:150], label="Actual", linewidth=2)
plt.plot(pred[:150], label="AIS+PSO Hybrid", linestyle='--', linewidth=2)
plt.title("⚖️ AIS+PSO Hybrid AquaHarvest Comparison Graph")
plt.xlabel("Samples")
plt.ylabel("Water Requirement")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_comparison_graph.png"))
plt.show()

# ============================================================
# 💾 Step 5: Save Results
# ============================================================
result = {
    "model_type": "Hybrid AIS + PSO",
    "selected_features": selected_features,
    "best_n_estimators": best_n_est,
    "best_max_depth": best_depth,
    "rmse": float(rmse),
    "r2_score": float(r2)
}

with open(os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_results.json"), "w") as f:
    json.dump(result, f, indent=4)
with open(os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_config.yaml"), "w") as f:
    yaml.dump(result, f)
joblib.dump(model, os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_model.pkl"))
joblib.dump({'scaler_x': scaler_x, 'scaler_y': scaler_y},
            os.path.join(OUTPUT_DIR, "Ais_Pso_hybrid_AquaHarvest_scalers.pkl"))

print("\n✅ [INFO] All outputs saved in:")
print(OUTPUT_DIR)
print("---------------------------------------------------")
print(" - Ais_Pso_hybrid_AquaHarvest_model.pkl")
print(" - Ais_Pso_hybrid_AquaHarvest_results.json")
print(" - Ais_Pso_hybrid_AquaHarvest_config.yaml")
print(" - Ais_Pso_hybrid_AquaHarvest_heatmap.png")
print(" - Ais_Pso_hybrid_AquaHarvest_prediction_graph.png")
print(" - Ais_Pso_hybrid_AquaHarvest_result_graph.png")
print(" - Ais_Pso_hybrid_AquaHarvest_comparison_graph.png")
print("---------------------------------------------------")
```
