```python
import os
import pandas as pd
import numpy as np
import json
import joblib
import yaml
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from geneticalgorithm import geneticalgorithm as ga
import warnings
warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# 🧩 Paths
# ------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Irrigation & Water Optimization"
files = [
    os.path.join(BASE_DIR, "archive", "plant_vase1(2).CSV"),
    os.path.join(BASE_DIR, "archive", "plant_vase1.CSV"),
    os.path.join(BASE_DIR, "archive", "plant_vase2.CSV")
]
os.makedirs(BASE_DIR, exist_ok=True)

# ------------------------------------------------------------
# 📊 Load and Merge Data
# ------------------------------------------------------------
dfs = []
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

data = pd.concat(dfs, ignore_index=True)
print("[INFO] Loaded and merged data shape:", data.shape)

# Normalize column names
data.columns = [col.strip().lower().replace(" ", "_") for col in data.columns]
data = data.dropna().reset_index(drop=True)

# ------------------------------------------------------------
# 🧼 Prepare Features & Target
# ------------------------------------------------------------
if 'timestamp' in data.columns:
    data['timestamp'] = pd.to_datetime(data['timestamp'], errors='coerce')
    data = data.sort_values('timestamp')

feature_cols = [c for c in data.columns if c not in ['timestamp', 'target', 'water_required']]
target_col = 'water_required' if 'water_required' in data.columns else feature_cols[-1]

X = data[feature_cols].values
y = data[target_col].values

# Scale inputs
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

# ------------------------------------------------------------
# 🧬 GA Optimizer (for hyperparameter tuning)
# ------------------------------------------------------------
def model_fitness(params):
    lstm_units = int(params[0])
    dropout_rate = params[1]
    lr = params[2]

    model = Sequential([
        LSTM(lstm_units, input_shape=(1, X_scaled.shape[1]), activation='tanh'),
        Dropout(dropout_rate),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr), loss='mse')

    model.fit(X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1]),
              y_scaled, epochs=3, batch_size=32, verbose=0)
    loss = model.evaluate(X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1]), y_scaled, verbose=0)
    return loss

varbound = np.array([[16, 128], [0.1, 0.5], [0.0001, 0.01]])

algorithm_param = {
    'max_num_iteration': 6,           # reduced for performance
    'population_size': 4,
    'mutation_probability': 0.2,
    'elit_ratio': 0.01,
    'parents_portion': 0.3,
    'crossover_probability': 0.5,
    'crossover_type': 'uniform',
    'max_iteration_without_improv': None  # ✅ fixed KeyError
}

print("[GA] Running Genetic Algorithm optimization...")
model_ga = ga(function=model_fitness, dimension=3, variable_type='real',
              variable_boundaries=varbound, algorithm_parameters=algorithm_param)
model_ga.run()

best_params = model_ga.output_dict['variable']
best_lstm, best_dropout, best_lr = int(best_params[0]), best_params[1], best_params[2]
print(f"[GA] Optimal Params → LSTM={best_lstm}, Dropout={best_dropout:.2f}, LR={best_lr:.5f}")

# ------------------------------------------------------------
# 🧠 Train Final LSTM Model
# ------------------------------------------------------------
model = Sequential([
    LSTM(best_lstm, input_shape=(1, X_scaled.shape[1]), activation='tanh'),
    Dropout(best_dropout),
    Dense(1)
])
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=best_lr), loss='mse')

print("[TRAINING] Training final hybrid LSTM model...")
history = model.fit(X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1]),
                    y_scaled, epochs=30, batch_size=32, verbose=1, validation_split=0.2)

# ------------------------------------------------------------
# 🔁 ARIMA component
# ------------------------------------------------------------
print("[ARIMA] Fitting ARIMA model...")
arima_model = ARIMA(y, order=(2, 1, 2))
arima_result = arima_model.fit()
arima_forecast = arima_result.predict(start=0, end=len(y)-1)

# ------------------------------------------------------------
# 🔗 Hybrid Output
# ------------------------------------------------------------
lstm_pred = scaler_y.inverse_transform(model.predict(X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])))
hybrid_pred = (0.6 * lstm_pred.flatten()) + (0.4 * arima_forecast)

# ------------------------------------------------------------
# 📈 Evaluation
# ------------------------------------------------------------
rmse = np.sqrt(mean_squared_error(y, hybrid_pred))
r2 = r2_score(y, hybrid_pred)
print(f"[RESULT] RMSE: {rmse:.4f} | R²: {r2:.4f}")

# ------------------------------------------------------------
# 💾 Save Outputs
# ------------------------------------------------------------
h5_path = os.path.join(BASE_DIR, "AquaHarvest_model.h5")
pkl_path = os.path.join(BASE_DIR, "AquaHarvest_scalers.pkl")
yaml_path = os.path.join(BASE_DIR, "AquaHarvest_config.yaml")
json_path = os.path.join(BASE_DIR, "AquaHarvest_results.json")

model.save(h5_path)
joblib.dump({'scaler_x': scaler_x, 'scaler_y': scaler_y}, pkl_path)

config = {
    'model_type': 'Hybrid LSTM + ARIMA + GA',
    'best_params': {'lstm_units': best_lstm, 'dropout': best_dropout, 'learning_rate': best_lr},
    'rmse': float(rmse),
    'r2': float(r2),
    'input_features': feature_cols,
    'target': target_col,
    'dataset_shape': data.shape
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f)

with open(json_path, "w") as f:
    json.dump(config, f, indent=4)

# ------------------------------------------------------------
# 📊 Visualization
# ------------------------------------------------------------
plt.figure(figsize=(9,5))
plt.plot(y, label="Actual", linewidth=2)
plt.plot(hybrid_pred, label="Hybrid Prediction", linestyle='--')
plt.title("AquaHarvest - Actual vs Predicted Water Requirement")
plt.xlabel("Samples")
plt.ylabel("Water Requirement")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Model Training Loss")
plt.legend()
plt.tight_layout()
plt.show()

print("\n✅ [INFO] All model files saved successfully in:")
print(BASE_DIR)
print("---------------------------------------------------")
print(" - AquaHarvest_model.h5")
print(" - AquaHarvest_scalers.pkl")
print(" - AquaHarvest_config.yaml")
print(" - AquaHarvest_results.json")
print("---------------------------------------------------")
```
